## probablility density functions
\begin{align}
\mbox{Gaussian} \\
\ g(y, \mu, \sigma) = \frac{1}{\sigma \sqrt{2 \pi}}  \exp{\frac{-(y - \mu)^2}{2 \sigma^2}} \\
\end{align}
## adaptable parameters
\begin{align}
\Delta \mu = \alpha_{\mu} (r - b{\mu}) \frac{y - \mu}{\sigma^2} \\
\\
\Delta \sigma = \alpha_{\sigma} (r - b{\sigma}) \frac{(y - \mu)^2 - \sigma^2}{\sigma^3} \\
\end{align}

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np

def g(y,u,s):
    c = 1 / (np.sqrt(2*np.pi) * s)
    return c * np.exp((-(y-u)**2)/(2*s**2))
```

```python
n_points = 1000
u = 6
s = 1
x = np.linspace(3, 9, n_points)
y = g(x,u,s)
plt.plot(x, y)
```
